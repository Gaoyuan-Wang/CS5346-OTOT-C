```python
# For file uploading
import requests
from io import BytesIO

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def data_upload(link):
  response = requests.get(link)
  content = response.content
  return pd.read_csv(BytesIO(content))
```

```python
# main dataset
df = data_upload('https://drive.google.com/uc?id=1fQYjaxUd7WYeGC3dSX4gKwv0MI8uWUwe&export=download')
```

```python
target_countries = ['Singapore', 'United States']
all = df[df['location'].isin(target_countries)]
singapore = df[df['location'].isin(['Singapore'])]
us = df[df['location'].isin(['United States'])]
```

```python
# singapore['date']
from matplotlib.pyplot import MultipleLocator
fig, ax = plt.subplots()

ax.plot(singapore['date'], singapore['new_cases'], label='new cases')
ax.xaxis.set_major_locator(MultipleLocator(180))
ax.set(xlabel='Date', ylabel='Case Number', title='New Cases Number')
ax.grid()
ax.legend()
plt.xticks(rotation=90)
plt.gcf().set_size_inches(20, 10)
plt.show()
```
